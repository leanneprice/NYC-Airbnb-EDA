Data Preprocessing and Feature Engineering 

```python
# Check NaN values in each column.
for column in listings.columns:
    print(column, listings[column].isnull().sum())
```

A lot of the columns with a high proportion of NaN values are descriptive such as 'neighbourhood_overview' (40% NaN) and 'house_rules' (45% NaN) and won't be used in this EDA. However, there are columns that would be useful to use in this EDA which have a high number of NaN values such as 'review_scores_rating' (30% NaN). 

One option is to use a subset list to remove rows that have an NaN value within a field of interest such as 'review_scores_rating', thereby preserving data in other rows. If .dropna() were used to drop a row containing an NaN in any column, a row would be removed if there was an NaN value in 'neighbourhood_overview', for example, but not 'review_scores_rating', thereby removing useful data. However, as there are a high proportion of NaN values within columns of interest such as 'review_scores_rating', removing rows containing NaN values from this column will also remove valuable data from other fields, 'bed' perhaps. As a lot of the numeric fields of interest contain no NaN values such as 'beds', 'latitude', 'longitude' etc. (see hypotheses). Seperate dataframes will be created to analyse fields of interest that have a high proportion of NaN values individually, such as review_scores_rating (see hypotheses.)  

It useful to remove a listing that has NaN values in all columns, so this will be performed in this first instance.


```python
# There are no rows with all columns as NaN
listings = listings.dropna(how="all") # 30,179 rows x 81 columns

# There are no duplicate rows
listings = listings.drop_duplicates() # 30,179 rows x 81 column
```

```python
# Dataframe for analysis of the host features of interest
host_data = listings[['price', 'host_since', 'host_response_time', 'host_response_rate', 'host_is_superhost']] # 30179 rows x 5 columns

# Remove a row if NaN value in any column
host_data = host_data.dropna() # 17082 rows x 5 columns
```

```python
# Dataframe for analysis of the review feature of interest
review_data = listings[['price','review_scores_rating']] # 30179 rows x 2 columns

# Remove a row if NaN value in any column
review_data = review_data.dropna() # 21094 rows x 2 columns
```

```python
# Dataframe for geo visualisation
geovis_data = listings[['price', 'latitude', 'longitude']] # 30179 rows x 3 columns
geovis_data = geovis_data.dropna() # 30179 rows x 3 columns
```
